# Exploratory Data Analysis

## Univariate Analysis

This analysis is only one column of Data. We are going to analyze only data of one columns and not in relation to any other column.

This can be done on both Numerical and Categorical Data.

```python
# using the titanic dataset fot demonstration
import pandas as pd
import seaborn as sns
df = pd.read_csv('train.csv')
# first ask those first 7 questions
```

```python
df.head()
```

### 1. Categorical Data
a. Countplot

Think of it in this way, what columns are needed for Survival? Some data is really correlated to survival and some are not.

```python
sns.countplot(x=df['Survived'])
# df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x=df['Pclass'])
```

```python
sns.countplot(x=df['Sex'])
```

b. PieChart

If you want a percentage comparision of all the possible values.

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
```

### 2. Numerical Data

It is different from Categorical Data where we are only concerned about count realted attributes of the data.

a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

b. Distplot

This is Histogram with a KDE.

```python
sns.distplot(df['Age'])
```

c. BoxPlot

This gives you the 5 number summary of any numerical data. The median Q1 to Q4 and the outliers.

- Min = Q1 - 1.5(IQR)
- Max = Q3 - 1.5(IQR)

```python
sns.boxplot(df['Fare'], orient='h')
# we can see that there are many outliers(potential)
```

```python
df['Age'].skew()
# if this is perfectly symmetrical this would be 1
```
